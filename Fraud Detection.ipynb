```python
## outlier function
def Outliers(col): 

    # calculating mean and std of the array
    mean, std = np.mean(col), np.std(col)
    limit = std * 3

    #Calculating the higher and lower cut values
    lower, upper = mean - limit, mean + limit

    # creating an array of lower, higher and total outlier values 
    outliers_lower = [x for x in col if x < lower]
    outliers_higher = [x for x in col if x > upper]
    outliers_total = [x for x in col if x < lower or x > upper]

    # array without outlier values
    outliers_removed = [x for x in col if x > lower and x < upper]
    
    print('total lower outliers: %d' % len(outliers_lower)) 
    print('total upper outliers: %d' % len(outliers_higher)) 
    print('Total outlier observations: %d' % len(outliers_total)) 
    print('Non-outlier observations: %d' % len(outliers_removed)) 
    print("Total percentual of Outliers: ", round((len(outliers_total) / len(outliers_removed) )*100, 4)) # Percentual of outliers in points
    
    return
```

```python
from sklearn.preprocessing import RobustScaler
rob_scaler = RobustScaler()

credit_data['amount'] = rob_scaler.fit_transform(credit_data['amount'].values.reshape(-1,1))
credit_data['time'] = rob_scaler.fit_transform(credit_data['time'].values.reshape(-1,1))
```

```python
from sklearn.model_selection import train_test_split
from imblearn.pipeline import make_pipeline 
from imblearn.over_sampling import SMOTE
```

```python
from sklearn.model_selection import KFold
```

```python
target=credit_data['class']
train=credit_data.drop('class',axis=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, accuracy_score, classification_report,precision_recall_curve,roc_curve,auc
```

```python
RFC_METRIC = 'gini'  #metric used for RandomForrestClassifier
NUM_ESTIMATORS = 100 #number of estimators used for RandomForrestClassifier
NO_JOBS = 4 #number of parallel jobs used for RandomForrestClassifier
```

```python
x_train,x_test,y_train,y_test= train_test_split(train,target,test_size=0.3,random_state=12)
kf=KFold(n_splits=5,random_state=6,shuffle=True)

accuracy_lst = []
precision_lst = []
recall_lst = []
f1_lst = []
auc_lst = []

rf_clf = RandomForestClassifier(n_jobs=NO_JOBS, 
                             random_state=12,
                             criterion=RFC_METRIC,
                             n_estimators=NUM_ESTIMATORS,
                             verbose=False)

for train_idx, val_idx in kf.split(x_train,y_train):
    pipeline = make_pipeline(SMOTE(sampling_strategy='minority'), rf_clf) 
    model = pipeline.fit(x_train.iloc[train_idx], y_train.iloc[train_idx])
    prediction = model.predict(x_train.iloc[val_idx])
    accuracy_lst.append(pipeline.score(x_train.iloc[val_idx], y_train.iloc[val_idx]))
    precision_lst.append(precision_score(y_train.iloc[val_idx], prediction))
    recall_lst.append(recall_score(y_train.iloc[val_idx], prediction))
    f1_lst.append(f1_score(y_train.iloc[val_idx], prediction))
    auc_lst.append(roc_auc_score(y_train.iloc[val_idx], prediction))
    
print('---' * 30)
print('')
print("accuracy: {}".format(np.mean(accuracy_lst)))
print("precision: {}".format(np.mean(precision_lst)))
print("recall: {}".format(np.mean(recall_lst)))
print("f1: {}".format(np.mean(f1_lst)))
print('---' * 30)
```

```python
# test data
labels = ['No Fraud', 'Fraud']
test_prediction = model.predict(x_test)
print(classification_report(y_test, test_prediction, target_names=labels))
```

```python
from sklearn.metrics import confusion_matrix
rf_cm=confusion_matrix(y_test, test_prediction)
sns.heatmap(rf_cm, annot=True, cmap='YlGnBu')
```

```python
def plot_roc_curve(clf, x_train, y_train):
    
    y_pred_score = clf.predict_proba(x_train)[:,1]
    #print "y_pred_score: "
    #print(y_pred_score)
    #print "y_true: "
    #print(y_train.values.ravel())
    fpr, tpr, thresholds = roc_curve(y_train.values.ravel(), y_pred_score)
    roc_auc = auc(fpr,tpr)
    # Plot ROC
    plt.figure(figsize=(10,6))
    plt.title('Receiver Operating Characteristic')
    plt.plot(fpr, tpr, 'b',label='AUC = %0.2f'% roc_auc)
    plt.legend(loc='lower right')
    plt.plot([0,1],[0,1],'r--')
    plt.xlim([-0.1,1.0])
    plt.ylim([-0.1,1.01])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
plot_roc_curve(rf_clf, x_train, y_train)
```
